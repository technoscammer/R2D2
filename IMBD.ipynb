```python
#importing library functions 
from bs4 import BeautifulSoup
import requests
import re

# Downloading IMDB's datasets from website
url = 'https://www.imdb.com/chart/top'
file = requests.get(url)
Y = BeautifulSoup(file.text, 'lxml')

movies = Y.select('td.titleColumn')
links = [a.attrs.get('href') for a in Y.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in Y.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in Y.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in Y.select('td.ratingColumn strong')]

imdb = []

# Store each item into dictionary (data), then put those into a list (imdb)
for index in range(0,len(movies)):
    # Seperate movie into: 'place', 'title', 'star-cast','ratings'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],}
    imdb.append(data)

for item in imdb:
     print(item['place'], '-', item['movie_title'],item['star_cast'],'-',item['rating'])
  
```
